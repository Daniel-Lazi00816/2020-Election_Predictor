```python
polls = pd.read_csv('../Data/president_polls.csv', usecols=['state', 'start_date', 'answer', 'pct', 'pollster_id'], 
                    parse_dates= ['start_date'], infer_datetime_format='True')
```

```python
candidates = ['Biden', 'Sanders', 'Trump']
polls = polls[polls.answer.isin(candidates)]
```

Since last meeting I have decided to take a different approach and that'll be to conduct bayesian inference on participation rates in 2016 Demcratic primary election. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pymc3 as pm
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
df = pd.read_csv('../Data/18-29.csv')
```

```python
cols = ['18-29_2020', '18-29_2016']
y = pd.DataFrame(imp.fit_transform(df[cols]), columns = cols)
y['State'] = df['State'].copy()
```

```python
y
```

```python
with pm.Model() as model:
    # Set priors for parameters
    beta_0 = pm.Normal('beta_0', mu=0, sd=5)
    beta_1 = pm.Normal('beta_1', mu=1, sd=0.2)
    sigma = pm.Uniform('sigma', lower=0, upper=8)
    Y_obs = pm.Normal(
        'Y_obs',
        mu = beta_0 + beta_1*y['18-29_2016'],
        sd = sigma,
        observed = y['18-29_2020']
    )
    trace = pm.sample(10000, tune=1000)
```

```python
pm.traceplot(trace);
```
